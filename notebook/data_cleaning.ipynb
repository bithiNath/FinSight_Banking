f. fill missing values in each numeric column with that column's median:

```python
for col in ["Age", "Balance", "CreditScore"]: 
  df[col] = df[col].fillna(df[col].median())
```

g. print value counts for each column to check unique values and distribution:

```python
cols = ['CustomerID', 'Age', 'AccountType', 'Balance', 'OpenDate', 'Status', 'CreditScore', 'Region', 'TransactionCount', 'HasLoan']

for col in cols:
  print(f"----{col}----")
  print(df[col].value_counts())
```

h. hsndling error values: 

step-1: normalize all string columns to capitalize format for consistency:

```python
for col in df.select_dtypes(include='string').columns:
    df[col] = df[col].astype(str).str.capitalize()
```

step-2: Replace 'Unknown' in 'OpenDate' with pd.NA to mark missing dates

```python
df['OpenDate'] = df['OpenDate'].replace('Unknown', pd.NA)
```

i. final cross check:

```python
cols = ['CustomerID', 'Age', 'AccountType', 'Balance', 'OpenDate', 'Status', 'CreditScore', 'Region', 'TransactionCount', 'HasLoan']

for col in cols:
  print("f----------{col}---------")
  print(df[col].value_counts())
```
